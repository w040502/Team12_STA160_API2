```python
import joblib
from pathlib import Path

big = joblib.load("airbnb_ensemble_artifacts.pkl")
```

```python
small_artifacts = {
    "city_models": big["city_models"],
    "feature_columns": big["feature_columns"],
    "feature_medians": big["feature_medians"]
}
```

```python
import joblib

joblib.dump(
    small_artifacts,
    "airbnb_small_model.pkl",
    compress=3
)
```

```python
joblib.dump(
    small_artifacts,
    "airbnb_small_model.pkl",
    compress=('zstd', 9)   # very strong compression
)
```
